```python
# Pool QC           2917
# Misc Feature      2824
# Alley             2732
# Fence             2358
# Order
# PID

cols_drop = ["Pool QC", "Misc Feature", "Alley", "Fence", "Order", "PID"]
df = df.drop(columns=cols_drop)
```

```python
df.info()
```

# Filling the missing values

```python
# IN categorical data we will be filling none

cat_fill_none = ["Mas Vnr Type", "Fireplace Qu", "Garage Type", "Garage Finish", "Garage Qual", "Garage Cond", "Bsmt Qual", "Bsmt Cond", "Bsmt Exposure",
                "BsmtFin Type 1", "BsmtFin Type 2"]
for col in cat_fill_none:
    if col in df.columns:
        df.fillna({col : 'None'}, inplace= True)
```

```python
df.isnull().sum().sort_values(ascending= False).head(20)
```

```python
# Numerical data we will be filling median value

num_fill_median = ["Lot Frontage", "Mas Vnr Area", "Garage Yr Blt"]
for col in num_fill_median:
    if col in df.columns:
        df.fillna({col:df[col].median()}, inplace= True)
```

```python
df.isnull().sum().sort_values(ascending= False).head(20)
```

```python
# fill the data with 0

Zero_fill = ["Bsmt Half Bath", "Bsmt Full Bath", "BsmtFin SF 2", "Garage Cars", "Garage Area", "BsmtFin SF 1", "Bsmt Unf SF", "Total Bsmt SF"]

for col in Zero_fill:
    if col in df.columns:
        df.fillna({col:0},inplace=True)
```

```python
df.isnull().sum().sort_values(ascending= False).head(20)
```

```python
# Electrical is object type data

if "Electrical" in df.columns:
    df["Electrical"]=df["Electrical"].fillna(df["Electrical"].mode())
```
